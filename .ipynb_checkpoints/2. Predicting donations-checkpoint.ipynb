# Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns', None)
```

```python
from sklearn import linear_model
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score, mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.ensemble import RandomForestRegressor
```

# Functions

```python
# A function to test different models on our dataframe

def tester(X_train, y_train, X_test, y_test, models, modelnames):
    # Iterating over all models
    for i in range(len(models)):
        model = modellist[i]
        # Fitting
        model.fit(X_train, y_train)
        # Predicting
        predictions = model.predict(X_test)
        # R2-Score
        r2scoret = round(r2_score(y_test, predictions),2)
        # Mean Square Error
        mset = round(mean_squared_error(y_test,predictions),2)
        # Root Mean Square Error
        rmset = round(np.sqrt(mset),2)
        # Mean Absolute Error
        maet = round(mean_absolute_error(y_test, predictions),2)
        # Giving out results:
        print(str(modelnames[i]) + ' results:')
        df_val = pd.DataFrame({'ValType': ['Test'], 'R2-Score': [r2scoret], 'MSE': [mset] , 'RMSE': [rmset], 'MAE': [maet]})
        display(df_val)
```

```python
# Reusing the function from the previous workbook, to check multicollinearity

def corr_check(model, df_input, number_of_columns):
    # Making a dataframe of the scores and column names
    df = pd.DataFrame(data = model.scores_, columns = ['score'])
    df['Column'] = df_input.columns
    # Sort it
    df_sorted = df.sort_values(by = ['score'], ascending = False).reset_index()
    # Making a list of the first x columns
    collist = []
    for i in range(number_of_columns):
        collist.append(df_sorted['Column'][i])
    # creating a correlation matrix
    correlations_matrix = df_input[collist].corr()
    # create a heatmap of it
    plt.figure(figsize = (16,16))
    sns.heatmap(correlations_matrix, annot=True, fmt='.2f')
    plt.show()
    return collist
```

# Loading File

```python
df = pd.read_csv('encoded_predicted.csv')
df.shape
df.head(1)
```

```python
# For building the model, we need only those who donated
df_donors = df[df['TARGET_B']==1].copy()
```

# Removing unwanted columns

The column target B is now all 1's and of no value for the model. 

The column Predicted_B is a result of our previous model and can be generated for new data, therefore we can keep it and see if it is useful. 

```python
df_build = df_donors.drop(['TARGET_B'], axis = 1).copy()
print(df_build.shape)
```

# X-y Split

```python
X = df_build.drop('TARGET_D',axis = 1)
```

```python
y = df_build['TARGET_D']
```

# Train-Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

# Testing model on raw data

```python
modellist = [
    KNeighborsRegressor(n_neighbors=3),
    linear_model.LinearRegression(),
    DecisionTreeRegressor(max_depth=3),
    RandomForestRegressor(max_depth=3)  ]
modelnames = ['KNN','Linear Regression','Decision Tree','Random Forest']
```

```python
tester(X_train, y_train, X_test, y_test, modellist, modelnames)
```

Looks like there is some work to do, but the Random Forest regressor and decision tree performed surprisingly well considering
the state of the data.

# Feature Selection

#### Variance selection
First we remove all columns with too low of a variance, since they cant possibly help us in a meaningful way.

```python
from sklearn.feature_selection import VarianceThreshold
# We set the variance threshhold at 0.02, we could test lower ones later,
# but we have a lot of columns to eliminate.
var_threshold = 0.02
sel = VarianceThreshold(threshold=(var_threshold))
```

```python
sel = sel.fit(X_train)
```

```python
# We create a temporary dataframe with the dropped columns
X_train_temp = pd.DataFrame(sel.transform(X_train), index = X_train.index)
X_test_temp = pd.DataFrame(sel.transform(X_test), index = X_test.index)
print(X_train_temp.shape)
print(X_test_temp.shape)
```

```python
# Testing if something imrpoved:
tester(X_train_temp, y_train, X_test_temp, y_test, modellist, modelnames)
```

After running it a few times, we see that the resilts improved a bit and got more consistent

#### Taking out columns with low variance

We now know that we have 237 columns with a high enough variance, we will continue with a dataframe just caontaining these columns.

```python
# Making a list of the columns we want:

# Putting the variances,  which are in the same order as the columns in the dataframe into a
# dataframe column and making a scond column containing the column names of the original dataframe
df_var = pd.DataFrame(data = sel.variances_, columns = ['variance'])
df_var['Column'] = X_train.columns
# Sort it
df_sorted = df_var.sort_values(by = ['variance'], ascending = False).reset_index()
# Making a list of the first 237 columns
collist = []
for i in range(237):
    collist.append(df_sorted['Column'][i])
```

```python
X_train_selected = X_train[collist]
X_test_selected = X_test[collist]
print(X_train_selected.shape)
print(X_test_selected.shape)
```

```python
# Testing out the lower variance dataframe
tester(X_train_selected, y_train, X_test_selected, y_test, modellist, modelnames)
```

The consistency of the resulst improved

# Conclusion

#### Calculating how high the prdicted income is
We will sum up the amount predicted. 
Since we now that we overpredict the amount of people, who will donate we will correct our predictions by that factor.

```python
# Sum of predicted donations:
sum_pred_d = round(df['Predicted_D'].sum())
sum_pred_d
```

```python
# Ratio of overpediction according to last mailing action:
ratio = df['Predicted_B'].sum()/df['TARGET_B'].sum()
ratio
```

```python
# Expected Donations:
exp_d = round(sum_pred_d/ratio)
exp_d
```

```python
# Expected expenses are the amount of predicted mailings times 0.68 for the cost per mail
exp_e = round(df['Predicted_B'].sum() * 0.68)
exp_e
```

```python
# We deduct the costs from the revenue to get the profit
exp_p = exp_d - exp_e
exp_p
```
